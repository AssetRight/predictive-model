```python
df_setor = pd.DataFrame(setor, columns=['Label','Id'])
df_setor.sort_values('Id', ascending = True)
```

```python
sectores_data = {}
for i in range(1,44):
  data = fundamentus.list_papel_setor(i)
  nome_setor = df_setor[df_setor['Id'] == i]['Label'].values[0]
  sectores_data[nome_setor] = data
sectores_data
```

```python
setor_codigo_map = {codigo: setor for setor, codigos in sectores_data.items() for codigo in codigos}
data = setor_codigo_map

# Criar uma série de pandas com base no dicionário
setor_series = pd.Series(data, name='Setor')

# Adicionar a série de setor ao DataFrame existente
df_with_sector = df.join(setor_series, on='papel')

# Visualizar o DataFrame atualizado
df_with_sector['Setor']
#df_with_sector = df_with_sector.dropna(subset=['Setor'])
df_with_sector = df_with_sector.sort_values(by="Setor")


df_with_sector




```
